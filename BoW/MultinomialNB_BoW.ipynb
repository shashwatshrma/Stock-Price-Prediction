```python
import pandas as pd
df = pd.read_csv('../train.csv')
df = df.dropna()
```

## Data Preprocessing

```python
train = df[df['Date'] < '2015-08-01']
test = df[df['Date'] > '2014-07-31']
print(train.shape)
print(test.shape)
train.head(10)
```

```python
# selecting independant variables
data=train.iloc[:,2:27]

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
wl = WordNetLemmatizer()
```

```python
# preprocessing the training data and creating a corpus of headlines

headlines = []

# converting the data to lowercase
for index in new_Index:
    data[index]=data[index].str.lower()

for row in range(0, len(data.index)):
    words = str(data.iloc[row,0:25])
    words = re.sub("[^a-zA-Z]"," ", words) # removing punctuations
    words = words.split()
    r = [ wl.lemmatize(word) for word in words if not (word in set(stopwords.words('english')) or word == 'b' or word == "b") ]
    r = ' '.join(r)
    headlines.append(r)
headlines[0]
```

## Implementing Bag of Words and Multinomial Naive-Baye's

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
# implement Bag of Words
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
# implement Multinomial Naive-Baye's
MNB_classifier=MultinomialNB()
MNB_classifier.fit(traindataset, train['Label'].values)
```

## Predicting using the test set and checking accuracy

```python
# preprocessing the testing data and creating a corpus of headlines similar to training set

test_data = test.iloc[:,2:27]
test_data.columns= new_Index

for index in new_Index:
    test_data[index]=test_data[index].str.lower()

test_transform= []
for row in range(0,len(test.index)):
    words = str(test_data.iloc[row,0:25])
    words = re.sub("[^a-zA-Z]"," ", words)
    words = words.split()
    r = [ wl.lemmatize(word) for word in words if not word in set(stopwords.words('english')) ]
    r = ' '.join(r)
    test_transform.append(r)
```

```python
# predicting using testing dataset

test_dataset = countvector.transform(test_transform)
predictions = MNB_classifier.predict(test_dataset)
```

```python
# to check accuracy

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
