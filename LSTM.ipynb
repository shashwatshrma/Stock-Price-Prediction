```python
import pandas as pd
train = pd.read_csv('./train.csv', encoding='ISO-8859-1')
train = train.dropna()
```

## Data Preprocessing

```python
# Get the Independent Features
X = train.drop('Label', axis=1)
X = X.drop('Date', axis=1)
```

```python
# Get the Dependent features
y = train['Label']
```

```python
X.head()
```

```python
X.shape
```

```python
y.shape
```

```python
new_Index = [str(i) for i in range(25)]
X.columns = new_Index
```

```python
messages = X.copy().reset_index()
messages = messages.drop('index', axis=1)
messages.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
# Text Preprocessing - removal of non-english characters, tokeization, and stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    words = str(messages.iloc[i])
    review = re.sub('b[\'\"]', '', words)
    review = re.sub('Name: ., dtype: object', '', review)
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

### Onehot Representation

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
```

```python
maximum=0
for sent in corpus:
    maximum=max(maximum, len(sent.split()))
maximum
```

```python
# Hyper parameters
voc_size = 10000
embedding_vector_features = 128
sent_length = 100
```

```python
onehot_repr = [one_hot(words,voc_size) for words in corpus] 
onehot_repr
```

### Embedding Representation

```python
embedded_docs = pad_sequences(onehot_repr, padding='post', maxlen=sent_length)
print(embedded_docs)
print(sent_length)
```

```python
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape, y_final.shape
```

## Creating and Training Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
```

```python
## Creating model
model=Sequential()
model.add(Embedding(input_dim = voc_size, output_dim = embedding_vector_features, input_length=sent_length))
model.add(LSTM(16))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

### Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, shuffle=False)
X_train.shape, X_test.shape
```

### Model Training

```python
# Finally Training
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=8,
          callbacks=EarlyStopping(patience=5, start_from_epoch=10, restore_best_weights=True))
```

## Prediction, Performance Metrics And Accuracy

```python
prediction = model.predict(X_test)
prediction
```

```python
# rounding up all the results because the metrics need whole numbers
y_pred=[]
for pred in prediction:
    y_pred.append(round(pred[0]))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
score = accuracy_score(y_test, y_pred)
print(score)
report = classification_report(y_test, y_pred)
print(report)
```
