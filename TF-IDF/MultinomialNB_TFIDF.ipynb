```python
import pandas as pd
df=pd.read_csv('../train.csv')
df.shape
```

## Data Preprocessing

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
test.head(10)
```

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
headlines[100]
```

## Implementing TF-IDF and Multinomial Naive-Baye's

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
## implement TF-IDF
tfidf=TfidfVectorizer(ngram_range=(2,2))
traindataset=tfidf.fit_transform(headlines)
```

```python
# implement Multinomial Naive-Baye's
MNB_classifier=MultinomialNB()
MNB_classifier.fit(traindataset, train['Label'].values)
```

## Predicting using the test set and checking accuracy

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = tfidf.transform(test_transform)
predictions = MNB_classifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
